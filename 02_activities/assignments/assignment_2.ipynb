### Bad visualization:

This data visualization illustrates the popularity of two political leaders prior to the 2022 presidential elections in Colombia.

Misleading Axes:

- One of the biggest problems here is that each leader has their own vertical axis, making it really hard to compare them directly. Petros popularity ranges from 25% to 37%, while Ficos goes from 3% to 19%. This makes Ficos increase look much more impressive than it actually is.
- If both candidates were shown on the same y-axis, wed get a much more honest comparison.

Line Representation Problems:

- The steep slope of Fico Gutierrezs line gives the impression of a dramatic rise in popularity, even though his actual numbers are quite low compared to Petro's. Meanwhile, Petros growth is shown as a relatively flat line, downplaying his increase.
- The way these lines are represented ends up making Ficos gain look far more significant than it is in reality.

Color and Readability Issues:

- The bright blue background with similarly-colored text makes the information really tough to read. The colors for the data points also blend in too much, making it harder for viewers to follow.
The design just doesnt do a good job of making the data stand out.

Lack of Context:

- There are no clear y-axis labels or an overall title explaining whats being shown. Plus, we dont see anything about the sample size or the margin of error for these polls, which is super important for understanding how reliable this data might be.


#### How could this data visualization have been improved?

Use a Shared Y-Axis:

- Instead of giving each leader their own y-axis, they should be using the same one. This would allow people to easily compare their popularity and get a more realistic sense of whats going on.

Change the Chart Type or Line Style:

- A shared bar chart might work better than a line chart here, since bar charts are generally better for comparing values between categories. If sticking with lines, they should at least be scaled properly to avoid giving a misleading impression of growth.

Fix the Color Scheme:

- The colors need to be more contrasting to improve readability. A lighter background with darker text would make things much easier to see. And the data points should stand out clearly so viewers can follow along without getting lost.

Add Contextual Information:

- Include things like sample size, data source, and margin of error. This gives viewers the context they need to understand how reliable these trends are.

- Adding a simple legend or label to explain what each line means would also make a big difference in helping people quickly grasp the point of the visualization.

Clarify the Timeline:

- Make sure the specific dates for each poll are clearly marked. That way, its easier to understand exactly when the changes in popularity happened.



### Good visualization:



This visualization presents the challenges faced by first-time home buyers in the United Kingdom from 1992 to 2023. It effectively communicates the relationships between house prices, incomes, and the amount of money borrowed by first-time buyers.

Clear and Easy to Follow:

- Simple Labels and Title: The title and labels are straightforward, making it easy to know whats being shown. The annotations explain trends in simple language, making the data accessible even for non-experts.

Great Use of Visual Elements:

- Color Coding: The blue and red lines clearly differentiate house prices and income, so you can follow the trends without confusion.
Highlighting Key Metrics: Important numbers like "5.4x" and "3.3x" are bold and easy to spot, ensuring viewers grasp the main points instantly.
- Clean Design: There are no unnecessary design elements, just a simple line chart that communicates the trends effectively.

Logical Storytelling:

- Logical Flow: The visualization moves logically from house prices to income to borrowing, helping viewers understand why home buying has become more challenging.
- Trends Over Time: The line chart effectively shows the progression and key shifts, making it easy to follow changes over the years.

Focused Presentation:

- The chart is efficienteverything included adds value, without excess decoration.

#### How could this data visualization have been improved?

Add Context:

- Including notes on major economic events, like the 2008 financial crisis, could help explain trends more clearly.

Interactivity:

- An interactive version where users could hover over points for exact values over time, would make it more engaging.

Reference Lines:

- Adding reference lines at key points would help viewers quickly understand significant changes.
Overall, the visualization is clear, well-designed, and tells a logical story. 


These small tweaks could enhance the depth and interactivity, but its already effective at communicating the challenges faced by first-time home buyers in a simple, impactful way.
