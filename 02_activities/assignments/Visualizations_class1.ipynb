```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

Making a sample 

```python
np.random.seed(613)

# Define the axis
x = np.arange(50)
y = np.random.randint(0,100,50) # 0 is the minimum, 100 is the maximum, 50 is how many values we generate  

fig,ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)  # Bar plot of x and y
```

```python
fig, ax =plt.subplots(figsize =(5,3))
ax.hist(y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years sice start')
fig.tight_layout()
```

```python
#Modifying lables

font1 ={'Family' : 'sans-serif','color':'blue','size':20}
font2 ={'family' : 'monospace','color':'green','size':14}
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years sice start', fontdict = font2)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```
